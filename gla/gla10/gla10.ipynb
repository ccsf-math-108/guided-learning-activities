### Measures of Centrality and Variability

* It is impossible to summarize data with a single value.
* Every summary involves some loss of information.
* It is common to describe the distribution from two perspectives:
    * Centrality Measures – describe the center of the data.
        * Examples: Mean, Median, Mode
    * Variability Measures – describe how the data is spread.
        * Examples: Range, Interquartile Range, Variance, Standard Deviation, Number of Categories

---

#### Examples Metrics

| **Type** | **Measure**                   | **Description**                                                       | **Recommended Use**                                                            |
| ------------ | ----------------------------- | --------------------------------------------------------------------- | -------------------------------------------------------------------------- |
| **Center**   | **Mean or Average**            | Sum of all values divided by the number of values.                    | When data are roughly symmetric and have no extreme outliers.              |
| **Center**   | **Median**                    | Middle value when data are ordered.                                   | When data are skewed or have outliers.                                     |
| **Center**   | **Mode**                      | Most frequently occurring value.                                      | When data are categorical or have repeated values.                         |
| **Spread**   | **Range**                     | Difference between maximum and minimum values.                        | Quick sense of total spread; sensitive to outliers.                        |
| **Spread**   | **Interquartile Range (IQR)** | Difference between the 75th and 25th percentiles (Q3 − Q1).           | When comparing variability of skewed data or data with outliers.           |
| **Spread**   | **Variance**                  | Average squared deviation from the mean.                              | When analyzing data mathematically (used in formulas).                     |
| **Spread**   | **Standard Deviation (SD)**   | Square root of the variance; measures average distance from the mean. | When data are roughly symmetric and you want spread in same units as data. |
| **Spread**   | **Number of Categories**   | The number of categories. | When data are categorical. |


---

#### Properties of Means

- From the [MATH 108 textbook](https://ccsf-math-108.github.io/textbook/chapters/14/1/Properties_of_the_Mean.html):
    - It need not be an element of the collection.
    - It need not be an integer even if all the elements of the collection are integers.
    - It is somewhere between the smallest and largest values in the collection.
    - It need not be halfway between the two extremes; it is not in general true that half the elements in a collection are above the mean.
    - If the collection consists of values of a variable measured in specified units, then the mean has the same units too.
- The mean value is the value which equializes, balances the entire distribution. 
- A proportion is a mean.
- In relation to the medain:
    - The mean and median are equal in a perfectly symetric distribution.
    - The mean is larger (smaller) than the median in a distribuiton that is skewed right (left).

---

### Chebychev’s Bound

- For **all** collections of numerical values and all numbers $z$, the proportion of entries that are in the range “average $\pm z$ SDs” is at least $1 - \frac{1}{z^2}$.
- Any numerical data value can be described in standard unit as $z = \frac{\text{value} - \text{average}}{\text{SD}}$.

### The Normal Distribution

---

#### Overview

- The normal curve appears a lot in data/statistical analysis.
- Each normal curve is uniquely defined by its mean $\mu$ and standard deviation $\sigma$.
- The standard normal curve has a mean value of 0 and a standard deviation of 1.
- There are many situations where a histogram's outline can be approximated with a normal curve.
- Like a histogram height, the densities (heights) associated with a normal curve is calculated by `stats.norm.pdf` from the `scipy` library.
- Like the areas of the bars in a histogram, the area under the the normal curve is calculated by `stats.norm.cdf` from the `scipy` library.
- The inflection points on the normal curve help identify 1 standard deviation away from the mean (center).

### The Central Limit Theorem

- How does the Normal distribution relate to summary statistics like averages and other sums?
- Regardless of the distribution of a variable, the probability distribution of the sum or average of a large random sample drawn with replacement will be roughly normal.
- For any population, the distribution of very possible sample average can be described b:
    - The mean of the probability distribution for a sample proportion is the mean of the population.
    - The standard deviation of the probability distribution for a sample average is the population standard deviation divided by the square root of the sample size.
- This means that you can create a confidence interval approxiamation for a mean using $$\mu \pm z \frac{\sigma}{\sqrt{n}}$$                                                                 
    - $n$ is the sample size
    - $\mu$ is usually approximated by the average of the same
    - $\sigma$ is estimated too, but that estiamation varies.
    - the confidence level is associated with the value of $z$ (e.g. $z=2$ for 95\% confidence)

---

### Sample Size

- A 95\% confidence interval estimate for the population mean can be expressed as $\mu \pm 2 \frac{\sigma}{\sqrt{n}}$.
- The width of the interval is $4 \frac{\sigma}{\sqrt{n}}$.
- If you want the width to be at most a certain value $W$ and the population is made of binary values, then the sample size needs to be at least $\left(4 \cdot \frac{0.5}{W}\right) ^2$.
